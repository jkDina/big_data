```
#Загрузить данные в pandas
import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
data = pd.read_csv('u.data', header=None, sep='\t', skiprows=0, names=r_cols)

g_cols = ['genre', 'index']
genre = pd.read_csv('u.genre', sep='|', names=g_cols)

i_cols = ['amount', 'type']
info = pd.read_csv('u.info', header=None, sep=' ', names=i_cols)

m_cols = ['movie_id', 
          'title', 
          'release_date', 
          'video_release_date', 
          'imdb_url', 
          'unknown',
          'action',
          'adventure',
          'animation',
          'children\'s',
          'comedy',
          'crime',
          'documentary',
          'drama',
          'fantasy',
          'film-noir',
          'horror',
          'musical',
          'mystery',
          'romance',
          'sci-fi',
          'thriller',
          'war',
          'wester'
         ]
item = pd.read_csv('u.item', header=None, sep='|', encoding='latin-1', names=m_cols)

o_cols=['occupation']
occupation = pd.read_csv('u.occupation')

u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
user = pd.read_csv('u.user', sep='|', names=u_cols)
```

```
#Средствами pandasa соедините данные по оценкам с данными 
#по пользователям (u.user)
user_ext = user.merge(data, left_on='user_id', right_on='user_id')
print(user_ext)
```

```
#Построить модель, которая предсказывает оценку для фильма и пользователя 
#(воспользоваться RandomForestRegressor, тк мы решаем задачу регрессии)
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from datetime import date


df = user_ext.merge(item, left_on='movie_id', right_on='movie_id')


del df['unix_timestamp']
del df['movie_id']
del df['user_id']
del df['zip_code']
del df['video_release_date']
del df['imdb_url']
del df['title']

target = df['rating']
del df['rating']

df.sex = pd.factorize(df.sex)[0]
df.sex = pd.Categorical(df.sex)

date_map = {
    'jan': 0,
    'feb': 1,
    'mar': 2,
    'apr': 3,
    'may': 4,
    'jun': 5,
    'jul': 6,
    'aug': 7,
    'sep': 8,
    'oct': 9,
    'nov': 10,
    'dec': 11
}

def convert_date(x):
  try:
    if type(x) != str:
      return 0
    arr = x.lower().split('-')
    year = int(arr[2])
    month = date_map[arr[1]]
    day = int(arr[0])
    
    d = date(year, month + 1, day)
    
    ordinal = d.toordinal()
    return ordinal
  
  except Exception as err:
    print('err = ', err)
    return 0
  
df['release_date'] = df['release_date'].apply(convert_date)

df = pd.get_dummies(df, columns=['occupation'], prefix=['occupation'])

pd.set_option('display.max_columns', 500)

X_train, X_test, y_train, y_test = train_test_split(df, target)
model = RandomForestRegressor(max_depth=2, random_state=43, n_estimators=120)
model.fit(X_train, y_train)
```

```
#Оценить качество регрессора на отложенной (train_test_split) выборке по одной из метрик на выбор - RMSE, RMAE, R^2
import numpy as np
from sklearn.metrics import mean_squared_error
print(model.score(X_test, y_test))
predicts = model.predict(X_test)
print('mean_squared_error = ', mean_squared_error(predicts, y_test))
df1 = pd.DataFrame({'prediction': predicts, 'real': np.array(y_test)})
df1.prediction = df1.prediction.apply(lambda x: round(x))
print(df1)
```
