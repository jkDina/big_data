# Фильтры

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
# выбрать несколько столбцов

country_stats = data.filter(items = ['country', 'quantity'])
country_stats.head()
```

```python
average_level = data['quantity'].mean()
average_level
```

```python
data[data.quantity > average_level].head()
```

```python
# фильтр на подстроку
# найдем как называется Россия в этом датафрейме

data[ data['country'].str.contains('us', case=False) ]['country'].unique()
```

```python
print(data['country'].str)
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# () | (() | () & ())
filtered_countries = data[ (data['country']=='Russian Federation') | (data['country']=='Belarus') ]

filtered_countries.head(10)
```

```python
filtered_countries['country'].unique()
```

```python
data[ ['country', 'quantity'] ].head()
```

```python
# фильтры на строки

data.loc[1000:1005]
```

# Сортировка

```python
# Сортировка по столбцу

data.sort_values(by='quantity').head()
```

```python
# сортировка по убыванию

data.sort_values(by='quantity', ascending=False).head()
```

```python
# сортировка по нескольким столбцам

data.sort_values(by=['country', 'year', 'quantity'], ascending=[False, True, False]).head(50)
```

```python
#Используем transactions.csv.

#Для какой транзакции (столбец ID) были наибольшие расходы (столбец Cost) 
#в категории "_8" (столбец Product)?
```

```python
# Загрузить
df = pd.read_csv('transactions.csv')
df.head()
```

```python
# Профильтровать (по стоблцу Product, оставив только значение "_8")
df_filtered = df[df['Product']=='_8']
df_filtered.head()
```

```python
# Отсортировать по столбцу Cost по убыванию
df_filtered.sort_values(by='Cost', ascending=False)
```

# Группировки
