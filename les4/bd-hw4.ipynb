```python
#Загрузите датафрейм и выведите первые 10 строчек 
#Загружаем csv файл

data = pd.read_csv('ad_campaign.csv')
data.head(n=10)
```

```python
#Выведите минимальное значение столбца effect (продуктивность) для каждой компании
data1 = data.groupby('Название группы').min()[['Продуктивность']]
data1.head()
```

```python
#Упорядочите полученный датафрейм по этим значениям
data2 = data1.sort_values(by='Продуктивность', ascending=False)
data2
```

```python
#Постройте гистограмму компаний по этому рейтингу
ax = data2['Продуктивность'].hist()
ax.set_xlabel("Продуктивность")
ax.set_ylabel("Количество организаций")
```
